**All Functions and Their Uses**

```python
def index():
    return render_template('index.html') #Function to Find the HTML File in your Machine

@app.route('/encrypt', methods=['POST'])
def encrypt(): #Implements Encryption In HTML File
    message = request.form['message']
    encrypted_message = encrypt_data(message, cipher_suite)
    return render_template('index.html', output=f'Encrypted: {encrypted_message.decode("utf-8")}')

@app.route('/decrypt', methods=['POST'])
def decrypt(): #Implements DEcryption In HTML File
    encrypted_message = request.form['encrypted_message'].encode()
    decrypted_message = decrypt_data(encrypted_message, cipher_suite)
    return render_template('index.html', output=f'Decrypted: {decrypted_message}')

def encrypt_data(data, cipher_suite): #Function That Encrypts Messages
    encrypted_data = cipher_suite.encrypt(data.encode())
    return encrypted_data

def decrypt_data(encrypted_data, cipher_suite): #Function That DEcrypts Messages
    decrypted_data = cipher_suite.decrypt(encrypted_data).decode()
    return decrypted_data
```
